```python
import pandas as pd
import pickle
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

# ----------------------------------------------------------
# 1. Load trained model and fitted scaler
# ----------------------------------------------------------
import joblib

model = joblib.load("linear_regression.pkl")
scaler = joblib.load("lr_scaler.pkl")



# ----------------------------------------------------------
# 3. Select features used during training (same order!)
# ----------------------------------------------------------
features = ['yr', 'mnth', 'holiday', 'weekday', 'workingday',
            'weathersit', 'temp', 'windspeed']

X_new = df_new[features]

# ----------------------------------------------------------
# 4. Apply the SAME scaler used during model training
# ----------------------------------------------------------
X_new_scaled = scaler.transform(X_new)

# ----------------------------------------------------------
# 5. Make predictions
# ----------------------------------------------------------
y_pred = model.predict(X_new_scaled)

# Add predictions to your dataset
df_new["predicted_cnt"] = y_pred

# ----------------------------------------------------------
# 6. If dataset has actual cnt â†’ evaluate accuracy
# ----------------------------------------------------------
if "cnt" in df_new.columns:
    y_true = df_new["cnt"]

    mae = mean_absolute_error(y_true, y_pred)
    mse = mean_squared_error(y_true, y_pred)
    rmse = mse ** 0.5
    r2 = r2_score(y_true, y_pred)

    print("Model Evaluation on New Dataset:")
    print("MAE:", mae)
    print("MSE:", mse)
    print("RMSE:", rmse)
    print("R2 Score:", r2)

```
