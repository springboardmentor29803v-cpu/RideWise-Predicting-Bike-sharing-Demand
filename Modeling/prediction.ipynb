```python
import numpy as np
import joblib

# ==============================
# LOAD SAVED MODEL
# ==============================

MODEL_PATH = "decision_tree_model.pkl"

try:
    model = joblib.load(MODEL_PATH)
    print("\nğŸŒ³ Decision Tree Model Loaded Successfully!")
except:
    print("âŒ Error: Model file not found. Make sure 'decision_tree_model.pkl' is in the same folder.")
    exit()

# ==============================
# PREDICTION LOOP
# ==============================

def main():
    print("\nğŸš´ BIKE DEMAND PREDICTION USING DECISION TREE ğŸš´\n")

    while True:
        try:
            temp = float(input("ğŸŒ¡ Temperature: "))
            yr = int(input("ğŸ“… Year (0 = 2018, 1 = 2019): "))
            season = int(input("ğŸŒ¤ Season (1: Spring, 2: Summer, 3: Fall, 4: Winter): "))
            hum = float(input("ğŸ’§ Humidity: "))
            weathersit = int(input("â›… Weather Situation (1, 2, 3, 4): "))
            windspeed = float(input("ğŸƒ Wind Speed: "))
            mnth = int(input("ğŸ—“ Month (1â€“12): "))
            weekday = int(input("ğŸ“† Day of Week (0 = Sunday, 1 = Monday, â€¦ 6 = Saturday): "))
            holiday = int(input("ğŸ‰ Holiday? (0 = No, 1 = Yes): "))
            workingday = int(input("ğŸ¢ Working Day? (0 = No, 1 = Yes): "))
            weekend = int(input("ğŸ“Œ Weekend? (0 = No, 1 = Yes): "))

        except ValueError:
            print("âŒ Please enter valid numeric values!")
            continue

        # Prepare input in correct order
        features = np.array([[ 
            temp, 
            yr, 
            season, 
            hum, 
            weathersit, 
            windspeed, 
            mnth, 
            weekday, 
            holiday, 
            workingday, 
            weekend 
        ]])

        # Predict
        prediction = model.predict(features)[0]

        print(f"\nğŸ“Œ Predicted Bike Rentals: {round(prediction)}")
        print("=" * 40)

        again = input("Predict again? (y/n): ").lower()
        if again != 'y':
            print("Exiting...")
            break


if __name__ == "__main__":
    main()
```
