```python
data_path = r"D:\RideWise-Predicting-Bike-sharing-Demand\Data\day(new).csv"
```

```python
df_new = pd.read_csv("D:\RideWise-Predicting-Bike-sharing-Demand\Data\day(new).csv")
```

```python
# Drop unwanted columns
columns_to_drop = ['instant', 'casual', 'registered','dteday','atemp']
df_new = df_new.drop(columns=columns_to_drop)
print(f"Unwanted columns are: {columns_to_drop}")
```

```python
# Create weekend column (1 if weekend, 0 if not weekend)
df_new['weekend'] = ((df_new['weekday'] == 0) | (df_new['weekday'] == 6)).astype(int)
print("Created weekend column (1 for weekend, 0 for weekday)")
```

```python
import joblib

# Load models
lasso_model = joblib.load("lasso.pkl")
ridge_model = joblib.load("ridge.pkl")
elastic_net_model = joblib.load("Elastic.pkl")

# Load scaler
scaler = joblib.load("l,r,e_scaler.joblib")
```

```python
feature_columns = [
    'season', 'yr', 'mnth', 'holiday', 'weekday', 
    'workingday', 'weathersit', 'temp', 'hum', 'windspeed','weekend'
]
```

```python
X_new = df_new[feature_columns]
```

```python
X_new_scaled = scaler.transform(X_new)
```

```python
df_new["Lasso_Pred"] = lasso_model.predict(X_new_scaled)
df_new["Ridge_Pred"] = ridge_model.predict(X_new_scaled)
df_new["ElasticNet_Pred"] = elastic_net_model.predict(X_new_scaled)
```
