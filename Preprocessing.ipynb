Uploading and Inspecting the Data

```python
# ------------------------------------------
# STEP 1: Import necessary libraries
# ------------------------------------------
import pandas as pd
import numpy as np

# ------------------------------------------
# STEP 2: Load your dataset
# ------------------------------------------
df = pd.read_csv("/content/drive/MyDrive/Project/day.csv")   # your uploaded file location
print("Data Loaded Successfully!\n")

# ------------------------------------------
# STEP 3: View first few rows
# ------------------------------------------
print(df.head())

# ------------------------------------------
# STEP 4: Basic information about dataset
# ------------------------------------------
print("\nDataset Info:")
print(df.info())

# ------------------------------------------
# STEP 5: Check statistical summary
# ------------------------------------------
print("\nStatistical Summary:")
print(df.describe())

# ------------------------------------------
# STEP 6: Check for missing values
# ------------------------------------------
print("\nMissing Values:")
print(df.isnull().sum())
```

Preprocessing Steps

```python
df = df.drop(['instant', 'dteday', 'casual', 'registered'], axis=1)
```

```python
df = pd.get_dummies(df, columns=['season', 'mnth', 'weekday', 'weathersit'], drop_first=True)
```

```python
from sklearn.preprocessing import StandardScaler

num_cols = ['temp', 'atemp', 'hum', 'windspeed']
scaler = StandardScaler()
df[num_cols] = scaler.fit_transform(df[num_cols])
```

```python
df.to_csv("preprocessed_day.csv", index=False)
print("Preprocessed file saved as preprocessed_day.csv")
```
